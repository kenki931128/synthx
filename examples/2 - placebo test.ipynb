```python
import synthx as sx
```

### significant difference

```python
df = sx.sample(
    n_units=100,
    n_time=50,
    n_observed_covariates=3,
    n_unobserved_covariates=10,
    intervention_units=1,
    intervention_time=40,
    intervention_effect=1.02,
    noise_effect=0.2,
    scale=2,
    seed=42,
)
```

```python
dataset = sx.Dataset(
    df,
    unit_column = 'unit',
    time_column = 'time',
    y_column = 'y',
    covariate_columns = ['covariate_1', 'covariate_2', 'covariate_3'],
    intervention_units=1,
    intervention_time=40,
)
```

```python
effect_test, effects_placebo, sc_test, scs_placebo = sx.placebo_test(dataset)
```

```python
sc_test.plot()
```

```python
scs_placebo[1].plot()
```

```python
effect_test, effects_placebo
```

```python
sx.stats.calc_p_value(effect_test, effects_placebo)
```
